1. Uma mquina embala rao para cachorro em pacotes de 15 kg, mas cada pacote produzido acaba tendo um peso aleatrio, com distribuio Normal, com mdia 15kg, mas com desvio padro 0,1 kg. Obtenha a probabilidade de que um pacote selecionado aleatoriamente dessa produo tenha mais que 15,2 kg.

```python
p_1 = stats.norm.cdf(15.2,15,0.1)
```

Este trecho mostra qual a probabilidade de umpocote destes ter MENOS do que 15.2kg. Ou seja, seja p_1c a probabilidade de que tenha MAIS que 15.2kg, podemos dar p_1c como sendo:

```python
p_1c = 1-p_1
p_1c
```

2. Da mquina descrita acima, obtenha o percentil de ordem 95% para o peso dos pacotes. Ou seja, o valor tal que a probabilidade de um pacote ter peso menor que este seja de 95%.

```python
p_95 = stats.norm(15, 0.1).ppf(.95)
p_95
```

Portanto, podemos afirmar que, dadas as condies de hiptese da distribuio normal do peso dos pacotes, 95% dos pacotes tero peso menor ou igual a 15.164kg.

3. Suponha duas variveis aleatrias $Y_1$ e $Y_2$, a primeira com distribuio Normal Padro (mdia=0, varincia=1), a segunda com distribuio *t-studdent* com 5 graus de liberdade (tambm padro). Calcule, para cada uma, a probabilidade de ser menor que cada um dos valores contidos no objeto $Y$ na clula abaixo:

```python
Y=np.linspace(-6, 6, 200)
# Seu cdigo comea aqui
dist_tY1 = stats.t(5, 0, 1)
dist_nY2 = stats.norm(0, 1)

df_prob = pd.DataFrame(columns = ['Normal', 't-student'])

for n, c in zip(Y, range(0, len(Y))):
    df_prob.loc[c, 'Normal'] = dist_nY2.cdf(n)
    df_prob.loc[c, 't-student'] = dist_tY1.cdf(n)
df_prob
```

Ao observar nosso resultado, poderiamos facilmente pensar que h algo de errado, pelo fato de a parte normal ser nula em ambas as extremidades do df. No entanto, isso se trata na verdade de uma evidncia do que j deveria ser esperado, isto , o achatamento mais agressivo da funo normal do que da funo t com relao s bordas...

```python
df_prob['Normal'].value_counts()
```

4. Faa uma visualizao grfica dos dados do item anterior.

```python
sns.lineplot(x=Y, y=df_prob['Normal'], label = 'Normal')
sns.lineplot(x=Y, y=df_prob['t-student'], label = 't-student')
plt.show()
```

Como podemos ver, a distribuio t-student decai com menos velocidade, o que explica o porque seus valores no se tornam identicamente nulos dentro do intervalo avaliado, enquanto a normal, por sua vez, j se torna.
